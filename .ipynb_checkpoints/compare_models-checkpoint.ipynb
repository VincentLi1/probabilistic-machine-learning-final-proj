Present the results of the different models on the train, validate, and test sets

```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
MODELS = ["sgd", "swa", "abn_core", "abn_root_core"] #"swag", "bacs", "abn"]
colors = ["skyblue", "orange", "green", "blue"]
```

```python
results = []

for model in MODELS:
    with open(f"results/{model}_results.json") as f:
        results.append(json.load(f))
```

```python
print(results)
```

```python
labels = ["Brightness", "Contrast", "Defocus"]

filtered_results = [[results[i]["brightness-1"], results[i]["contrast-1"], results[i]["defocus_blur-1"]] for i in range(len(MODELS))]

x = np.arange(len(labels))
width = 0.2

# Plot
for i in range(len(MODELS)):
    plt.bar(x - width*((len(MODELS)-1)*0.5-i), filtered_results[i], width, label=MODELS[i], color=colors[i])

# Labels and formatting
plt.xlabel('Corruption (Intensity L1)')
plt.ylabel('Accuracy (Proportion)')
plt.title('Accuracy of Image Models on Corrupted Test Set\n (Train ImageNet, Test ImageNet-C)')
plt.xticks(x, labels)
plt.legend(title="Solver")
plt.tight_layout()
plt.show()
```
